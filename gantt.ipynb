```python
import matplotlib.pyplot as plt
import pandas as pd

# Lire les données à partir du fichier texte
df = pd.read_csv('Gantt ft10_9.txt', sep='\t')

# Définir une palette de couleurs pour chaque machine
couleurs_machine = {
    0: 'blue',
    1: 'green',
    2: 'red',
    3: 'purple',
    4: 'orange',
    5: 'yellow',
    6: 'cyan',
    7: 'magenta',
    8: 'lime',
    9: 'pink'
    # Ajoutez plus de machines avec leurs couleurs respectives si nécessaire
}

# Créer une figure et un axe
fig, ax = plt.subplots()

# Parcourir les lignes du dataframe et dessiner les barres
for index, row in df.iterrows():
    ax.barh(y=row['Job'], left=row['Début'], width=row['Fin']-row['Début'], align='center', color=couleurs_machine[row['Machine']], label=f'Machine {row["Machine"]}')

# Ajouter des étiquettes et des légendes
ax.set_xlabel('Temps')
ax.set_ylabel('Job')
ax.set_title('Diagramme de Gantt')
ax.invert_yaxis()

labels = [f"Machine {label}" for label in couleurs_machine.keys()]
# Créer une légende à partir du dictionnaire de couleurs des machines
handles = [plt.Rectangle((0,0),1,1, color=couleurs_machine[label]) for label in couleurs_machine]
# ax.legend(handles, couleurs_machine.keys())
# plt.legend(handles, couleurs_machine.keys(), bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left')

# Afficher le diagramme de Gantt
plt.show()
```
